```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Revenue per customer
customer_revenue = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
print(customer_revenue.head())
```

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False)
print(top_products.head())
```

```python
df.to_csv("OnlineRetail_Cleaned.csv", index=False)
```

```python
# 1 Filter transactions where Quantity > 1000 (possible outliers)
outliers = df[df['Quantity'] > 1000]
print("Suspicious outlier transactions:", outliers.shape)

# Optionally drop them if needed
df = df[df['Quantity'] <= 1000]

# 2 Sort by highest revenue orders
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Sort descending by Revenue
df_sorted = df.sort_values(by='Revenue', ascending=False)
print(df_sorted.head(10))
```

```python
# Create a sample product info DataFrame
product_info = pd.DataFrame({
    'StockCode': ['85123A', '71053', '84406B', '84029G', '85099B'],
    'Category': ['Home Decor', 'Toys', 'Kitchen', 'Home Decor', 'Accessories'],
    'Supplier': ['UK_Imports', 'ToyWorld', 'CookCraft', 'UK_Imports', 'GlamShop']
})

# Merge with main data
df_merged = pd.merge(df, product_info, on='StockCode', how='left')

print(df_merged[['StockCode', 'Description', 'Category', 'Supplier']].head())
```

```python
country_stats = df_merged.groupby('Country').agg(
    total_revenue=('Revenue', 'sum'),
    avg_revenue=('Revenue', 'mean'),
    order_count=('InvoiceNo', 'nunique')
).sort_values(by='total_revenue', ascending=False)

print(country_stats.head(10))
```

```python
# Create price range buckets
bins = [0, 2, 10, 50, 500, df['UnitPrice'].max()]
labels = ['Very Cheap', 'Cheap', 'Medium', 'Expensive', 'Luxury']

df_merged['Price_Range'] = pd.cut(df_merged['UnitPrice'], bins=bins, labels=labels, include_lowest=True)

# Group by price range
price_range_stats = df_merged.groupby('Price_Range').agg(
    total_purchases=('InvoiceNo', 'count'),
    avg_revenue=('Revenue', 'mean'),
    total_revenue=('Revenue', 'sum')
)

print(price_range_stats)
```

```python
corr_matrix = df_merged[['Quantity', 'UnitPrice', 'Revenue']].corr()
print(corr_matrix)
```

```python
# Rename columns
df_merged = df_merged.rename(columns={'StockCode': 'Product_ID', 'UnitPrice': 'Price'})

# Convert descriptions to lowercase
df_merged['Description'] = df_merged['Description'].str.lower()

# Filter products that contain a keyword (e.g., "glass")
glass_products = df_merged[df_merged['Description'].str.contains('glass', na=False)]
print(glass_products[['Product_ID', 'Description', 'Revenue']].head())
```
