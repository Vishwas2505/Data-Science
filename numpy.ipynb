```python
#universal functions
x=np.array([1,2,3])
y=np.array([3,4,5])

#arithematicoperations
print("Arithematical operations")
print(np.add(x, y))
print(np.subtract(x, y))
print(np.multiply(x, y))
print(np.divide(x, y))
print(np.power(x, y))
print(np.mod(x, y))

print("\nExponential operations")
#exponential
print(np.exp(x))
print(np.log(x))
print(np.log10(x))
print(np.log2(x))
print(np.expm1(x))

print("\ntrigonmetrical operations")
#trigonmetric
print(np.sin(x))
print(np.cos(x))
print(np.tan(x))
print(np.arcsin(x))
print(np.arccos(x))
print(np.arctan(x))
print(np.degrees(x))
print(np.radians(x))

#rounding
print("\nrounding operations")
print(np.round(x))
print(np.floor(x))
print(np.ceil(x))
print(np.trunc(x))

#statistical
print("\nstatistical operations")
print(np.min(x))
print(np.max(x))
print(np.mean(x))
print(np.std(x))
print(np.var(x))
print(np.sum(x))
print(np.prod(x))

#logical
print("\nlogical operations")
print(np.logical_and(x, y))
print(np.logical_or(x, y))
print(np.logical_not(x))
print(np.logical_xor(x, y))

#comparison
print("\ncomparsion operations")
print(np.greater(x, y))
print(np.less(x, y))
print(np.equal(x, y))
print(np.not_equal(x, y))
print(np.greater_equal(x, y))
print(np.less_equal(x, y))

#bitwise
print("\nbitwise operations")
print(np.bitwise_and(x, y))
print(np.bitwise_or(x, y))
print(np.bitwise_xor(x, y))
print(np.invert(x))
```

```python
farmA=np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC=np.array([[15,16,14],[14,13,15],[16,17,18,],[12,11,13]])
production_all=np.array([farmA,farmB,farmC])
print("shape",production_all.shape)
print("size:",production_all.size)
print("datatype",production_all.dtype)
print("dimensions",production_all.ndim)


#arithematic and aggregatic functions
total_per_farm = production_all.sum(axis=(1, 2))
print("total per farm",total_per_farm)
print("mean for farms",production_all.mean(axis=(1,2)))
print("std of farms",production_all.std(axis=(1,2)))
print("var of farms",production_all.var(axis=(1,2)))

#per hour mult idea
hour_weights=[0.5,1.0,0.75]
weighted_per_panel=np.dot(production_all,hour_weights)
print(weighted_per_panel)

#boardcasting
print("add after production b",production_all[1]+0.5)

#arithematic opeartions
mean_per_farm = production_all.mean(axis=(1, 2))
std_per_farm = production_all.std(axis=(1, 2), ddof=0)
mean_farm_sh = mean_per_farm.reshape(3,1,1)
std_farm_sh = std_per_farm.reshape(3,1,1)

normalized = (production_all - mean_farm_sh) / std_farm_sh
print("normalized shape:", normalized.shape)
print("normalized dtype:", normalized.dtype)
print("normalized first farm (farmA) first row:", normalized[0,0])
print()

#many ufuncs
eps = 1e-9
abs_norm = np.abs(normalized)

exp_norm = np.exp(normalized)
expm1_norm = np.expm1(normalized)
safe_log = np.log(abs_norm + eps)    # log(|x| + eps)
safe_log10 = np.log10(abs_norm + eps)
safe_log2 = np.log2(abs_norm + eps)

# Trigonometric:
sin_norm = np.sin(normalized)
cos_norm = np.cos(normalized)
tan_norm = np.tan(normalized)
# arcsin/arccos require inputs in [-1,1], clip to safe range
clipped_for_arcs = np.clip(normalized, -1.0, 1.0)
arcsin_norm = np.arcsin(clipped_for_arcs)
arccos_norm = np.arccos(clipped_for_arcs)
arctan_norm = np.arctan(normalized)

# degrees/radians example: convert arctan results to degrees
arctan_degrees = np.degrees(arctan_norm)
arctan_radians = np.radians(arctan_degrees)  # should match arctan_norm (roundtrip-ish)

# Rounding:
round_norm = np.round(normalized, 3)
floor_norm = np.floor(normalized)
ceil_norm = np.ceil(normalized)
trunc_norm = np.trunc(normalized)

print("Some ufunc summaries:")
print("exp mean:", np.mean(exp_norm))
print("log (safe) mean:", np.mean(safe_log))
print("sin mean:", np.mean(sin_norm))
print("round (first farm first row):", round_norm[0,0])
print()

#boardcast based on table
high_output = production_all > 20
even_hours_mask = np.array([True, False, True]).reshape(1,1,3)

# Combine masks:
high_and_even = np.logical_and(high_output, even_hours_mask)
high_or_even = np.logical_or(high_output, even_hours_mask)
high_not = np.logical_not(high_output)

print("high_output.any():", high_output.any())
print("high_and_even.any():", high_and_even.any())
print()

# Also use comparison ufuncs:
ge_6 = np.greater_equal(production_all, 6.0)
lt_5 = np.less(production_all, 5.0)
print("Number of readings >=6.0 kWh:", np.sum(ge_6))
print()
```
