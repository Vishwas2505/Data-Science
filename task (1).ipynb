```python
#2.trend analysis
plt.plot(df['total_bill'],df['time'],color='red',linestyle='--')
plt.xlabel('total_bill')
plt.ylabel('time')
plt.title("Trend Analysis")
plt.show()
```

```python
#2.trend analysis
plt.plot(df['tip'],df['time'],color='red',linestyle='--')
plt.xlabel('tip')
plt.ylabel('time')
plt.title("Trend Analysis")
plt.show()
```

```python
#3 Category Comparison
plt.bar(df['sex'],df['tip'],color='red')
plt.xlabel('sex')
plt.ylabel('tip')
plt.title("Category Comparison")
plt.show()
```

```python
#3 Category Comparison
plt.bar(df['sex'],df['total_bill'],color='red')
plt.xlabel('sex')
plt.ylabel('tip')
plt.title("Category Comparison")
plt.show()
```

```python
#4 valuedistribution
plt.hist(df['total_bill'],bins=10,color='red',edgecolor='black')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.title("Value Distribution")
plt.show()
```

```python
#5 relationship
plt.scatter(df['total_bill'],df['tip'],color='red')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title("Relationship")
plt.show()
```

```python
#6.ouliters&spread

plt.boxplot(df['total_bill'])
plt.xlabel('total_bill')
plt.title("Outliers & Spread")
plt.show()
```

```python
#7.propotions
plt.pie(df['sex'].value_counts(),labels=df['sex'].unique(),autopct='%1.1f%%')
plt.title("Proportions")
plt.show()
```

```python
#8.MULTIPLE VISUALS
plt.subplot(1,2,1)
plt.bar(df['sex'],df['tip'],color='red')
plt.xlabel('sex')
plt.ylabel('tip')
plt.title("Category Comparison")
plt.subplot(1,2,2)
plt.pie(df['sex'].value_counts(),labels=df['sex'].unique(),autopct='%1.1f%%')
plt.title("Proportions")
plt.show()
```
